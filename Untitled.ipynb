```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('property_listing_data_in_Bangladesh.csv')
```

```python
df.drop(['title','flooPlan','url','lastUpdated'],axis=1,inplace=True)
```

```python
df['area']=df['area'].str.split(' ').str[0]
```

```python
df['location']=df['adress'].str.split(',').str[1]
```

```python
df['City']=df['adress'].str.split(',').str[2]
```

```python
df.drop('adress',axis=1,inplace=True)
```

```python
df['price']=df['price'].str.split(' ').str[0]
```

```python

df['price'] = pd.to_numeric(df['price'])
```

```python
df['price']=df['price']*1000
```

```python
df['area'] = df['area'].str.replace(',', '')
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
df=pd.get_dummies(df, columns = ['type','City'])
```

```python
data=pd.DataFrame(df)
```

```python
data['City_ Sholokbahar'].value_counts
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['location']=le.fit_transform(data['location'])
```

```python
data['beds']=le.fit_transform(data['beds'])
```

```python
data['bath']=le.fit_transform(data['bath'])
```

```python
data.drop('purpose',axis=1,inplace=True)
```

```python
data.head(30)
```

```python
data['location'].value_counts()
```

```python
data=data*1
```

```python
data.head()
```

```python
y=data.iloc[:,3:4]
```

```python
X=data.drop('price',axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define the model
model = RandomForestClassifier(n_estimators= 100, min_samples_split=5, min_samples_leaf=1, max_depth=30, criterion='gini',bootstrap=False)

# Define parameter distributions
param_distributions = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
    'criterion': ['gini', 'entropy']  # Valid criteria for classifiers
}

# Create the RandomizedSearchCV object
rcv = RandomizedSearchCV(model, param_distributions=param_distributions, n_iter=10, scoring='accuracy', random_state=42)

# Fit the RandomizedSearchCV
rcv.fit(X_train, y_train)

# Access the best parameters
best_params = rcv.best_params_

# Display the best parameters
print("Best Parameters:", best_params)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
a=accuracy_score(y_test,y_pred)
```
